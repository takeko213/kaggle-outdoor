# ground_truth_train
各phoneのground_truthデータをマージして正解データセットを作る

```python
import os
import pandas as pd
import pathlib
from tqdm import tqdm
```

```python
# directory setting
INPUT = '../input/google-smartphone-decimeter-challenge'
OUTPUT = '../input/google-smartphone-decimeter-challenge/prep'
```

```python
p = pathlib.Path(INPUT)
gt_files = list(p.glob('train/*/*/ground_truth.csv'))
print('ground_truth.csv count : ', len(gt_files))
```

```python
dfs = []
for gt_file in tqdm(gt_files):
    dfs.append(pd.read_csv(gt_file))
df = pd.concat(dfs)
```

```python
df
```

```python
base_train = pd.read_csv(INPUT + '/' + 'baseline_locations_train.csv')
base_train
```

```python
# 並び順をbase_trainと合わせる
gt = base_train[['collectionName', 'phoneName', 'millisSinceGpsEpoch']].copy()
gt = gt.merge(df, on=['collectionName', 'phoneName', 'millisSinceGpsEpoch'], how='inner')
```

```python
gt.to_csv(OUTPUT + '/ground_truth_train.csv', index=False)
```
