# derived
各phoneのderivedデータをマージして正解データセットを作る

```python
import os
import pandas as pd
import pathlib
from tqdm import tqdm
```

```python
# directory setting
INPUT = '../input/google-smartphone-decimeter-challenge'
OUTPUT = '../input/google-smartphone-decimeter-challenge/prep'
```

### train

```python
p = pathlib.Path(INPUT)
gt_files = list(p.glob('train/*/*/*derived.csv'))
print('drived.csv count : ', len(gt_files))
```

```python
dfs = []
for gt_file in tqdm(gt_files):
    dfs.append(pd.read_csv(gt_file))
df = pd.concat(dfs)
```

```python
df
```

```python
base_train = pd.read_csv(INPUT + '/' + 'baseline_locations_train.csv')
base_train
```

```python
# 並び順をbase_trainと合わせる
gt = base_train[['collectionName', 'phoneName', 'millisSinceGpsEpoch']].copy()
gt = gt.merge(df, on=['collectionName', 'phoneName', 'millisSinceGpsEpoch'], how='inner')
```

```python
gt.to_csv(OUTPUT + '/derived_train.csv', index=False)
```

### test

```python
p = pathlib.Path(INPUT)
gt_files = list(p.glob('test/*/*/*derived.csv'))
print('derived.csv count : ', len(gt_files))
```

```python
dfs = []
for gt_file in tqdm(gt_files):
    dfs.append(pd.read_csv(gt_file))
df = pd.concat(dfs)
```

```python
df
```

```python
base_test = pd.read_csv(INPUT + '/' + 'baseline_locations_test.csv')
base_test
```

```python
# 並び順をbase_trainと合わせる
gt = base_test[['collectionName', 'phoneName', 'millisSinceGpsEpoch']].copy()
gt = gt.merge(df, on=['collectionName', 'phoneName', 'millisSinceGpsEpoch'], how='inner')
```

```python
gt.to_csv(OUTPUT + '/derived_test.csv', index=False)
```
