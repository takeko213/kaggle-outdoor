# derived
各phoneのderivedデータをマージして正解データセットを作る

```python
import os
import pandas as pd
import pathlib
from tqdm import tqdm
```

```python
# directory setting
INPUT = '../input/google-smartphone-decimeter-challenge'
OUTPUT = '../input/google-smartphone-decimeter-challenge/prep'
```

### train

```python
p = pathlib.Path(INPUT)
gt_files = list(p.glob('train/*/*/*derived.csv'))
print('drived.csv count : ', len(gt_files))
```

```python
dfs = []
for gt_file in tqdm(gt_files):
    dfs.append(pd.read_csv(gt_file))
df = pd.concat(dfs)
```

```python
df
```

```python
df.to_csv(OUTPUT + '/derived_train.csv', index=False)
```

### test

```python
p = pathlib.Path(INPUT)
gt_files = list(p.glob('test/*/*/*derived.csv'))
print('derived.csv count : ', len(gt_files))
```

```python
dfs = []
for gt_file in tqdm(gt_files):
    dfs.append(pd.read_csv(gt_file))
df = pd.concat(dfs)
```
